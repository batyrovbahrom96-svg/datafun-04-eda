```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("penguins")
df.head()

df.describe(include="all")

df.hist(figsize=(10,8))
plt.tight_layout()
plt.show()

df['flipper_length_mm'].hist()

df.rename(columns={'flipper_length_mm': 'flipper_length'}, inplace=True)
df['body_mass_kg'] = df['body_mass_g'] / 1000

sns.scatterplot(x='flipper_length', y='body_mass_kg', hue='species', data=df)
plt.title("Body Mass vs Flipper Length by Species")
plt.show()


```

## Insights
- Gentoo penguins tend to have higher body mass.
- Adelie penguins generally have smaller flipper lengths.
- The scatterplot shows species separation by body size.
